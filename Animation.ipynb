```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
rc('animation', html='html5')

fig = plt.figure()
ax = plt.axes(xlim=(0, 1), ylim=(-1.2, 1.2))

ax.set_xticks([0,0.5,1])
ax.set_xticklabels(['$0$','$0.5$','$1$'])
ax.set_yticks([-1.2, -0.6, 0, 0.6, 1.2])
ax.set_xlabel('$x$')
ax.set_ylabel('$y(x,t)$')

plt.close()

line, = ax.plot([], [], lw=2)
def init():
    line.set_data([], [])
    return (line,)

def animate(t):
    x = np.linspace(0, 1, 300)
    y=np.sin(2.0*np.pi*x)*np.exp(-0.1*t)
    line.set_data(x, y)
    return (line,)

tt = list(np.linspace(0, 40, 200))
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=tt, interval=20, blit=True)
anim
```
